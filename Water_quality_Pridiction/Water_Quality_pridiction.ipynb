```python
# Fill missing numerical values with mean
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python

# Fill missing numerical values with mean
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
df["ph_cat"] = pd.cut(df["pH"], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1, 2, 3, 4, 5])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Assume income_cat is a column in the dataset created from median_income
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(df, df["ph_cat"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
#lets Drop/Remove   the income _cat column
for sett in(strat_train_set,strat_test_set):
    sett.drop("ph_cat",axis=1,inplace=True)
```

```python
# Separate numeric and object columns
numeric_cols = sample.select_dtypes(include=['float64', 'int64']).columns

object_cols = sample.select_dtypes(include=['object']).columns

print("Numeric Columns:", numeric_cols)
print("Object Columns:", object_cols)
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_imputer = SimpleImputer(strategy='mean')  # You can use median if needed
sample[numeric_cols] = num_imputer.fit_transform(sample[numeric_cols])
```

```python
obj_imputer = SimpleImputer(strategy='most_frequent')  # Use mode
sample[object_cols] = obj_imputer.fit_transform(sample[object_cols])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
sample['Month'] = le.fit_transform(sample['Month'])
sample['Source'] = le.fit_transform(sample['Source'])
sample['Color'] = le.fit_transform(sample['Color'])
```
